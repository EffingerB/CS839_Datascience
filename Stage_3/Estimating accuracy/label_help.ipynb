```python
import pandas as pd
import csv
import sqlite3
import math

IMDB_data = pd.read_csv("IMDB.csv")
Meta_data = pd.read_csv("Metacritic.csv")


conn = sqlite3.connect('movies.db')
IMDB_data.to_sql("IMDB", conn, if_exists="replace", index=False)
Meta_data.to_sql("Metacritic", conn, if_exists="replace", index=False)
```

```python
TRUE = True #may need to change
FALSE = False #may need to change
results = []
with open('reduced_Job_Movie_apply_rules_ds.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    the_files = list(csv_reader)
    all_files = the_files[1:]
    for candidate in all_files:
        id1 = candidate[0]
        id2 = candidate[1]
        QUERY = """
        SELECT M.title as Meta_Title, I.title as IMDB_Title,  M.release_year as Meta_Release_Year, I.release_year as IMDB_Release_Year
        FROM Metacritic M, IMDB I
        WHERE I._id = """ + id1 + """ AND M._id =""" + id2 + """
        """
        df =list(pd.read_sql(QUERY, conn).loc[0])
        real_output = [id1,id2]
        if(df[2] == "TBA" or df[3] == "TBA"):
            print(df)
            correctness = input("is this a match?")
            real_output.append(correctness)
        elif(df[0] == df[1] and (float(df[2]) - float(df[3])) <= 1):
            real_output.append(TRUE) #might need to change to True
        elif(abs((float(df[2]) - float(df[3]))) > 1):
            real_output.append(FALSE)
        else:
            print(df)
            correctness = input("is this a match?")
            real_output.append(correctness)
        #print(df)
        results.append(real_output)
        
```
